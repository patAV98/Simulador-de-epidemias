# **Variación del porcentaje de propagación**

Propagación del 15%

![texto del enlace](https://media.giphy.com/media/KblE8wSQdLJaKWca3P/giphy.gif)

Propagación del 30%

![texto alternativo](https://media.giphy.com/media/cMKFq8BgbF4s3Xh8Tu/giphy.gif)

Propagación del 50%

![texto alternativo](https://media.giphy.com/media/RNPxKVVIMQNVAM4hdc/giphy.gif)

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

names = ["r10", "r15", "r30", "r50"]
labels = ["10%", "15%", "30%", "50%"]
source = []
for name in names:
    source.append(pd.read_csv("./epidemic-data-" + name + ".csv"))
a = np.array(list(range(0,len(source[0]["infections"]))))
x = a.astype(float)
```

```
for s, n in zip(source, labels):
    plt.plot(x, s["infections"], label="r="+n)

plt.title('No of infections')     
plt.legend()
plt.show()
```

```
for s, n in zip(source, labels):
    plt.plot(x, s["deaths"], label="r="+n)

plt.title('No of deaths')    
plt.legend()
plt.show()
```
