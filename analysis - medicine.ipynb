```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

names = ["m075", "m05", "m03", "data"]
labels = ["75%", "50%", "30%" , "none"]
source = []
for name in names:
    source.append(pd.read_csv("data/epidemic-" + name + ".csv"))
a = np.array(list(range(0,len(source[0]["infections"]))))
x = a.astype(float)

```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["infections"], label="m="+n)

plt.title('No of infections')    
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["deaths"], label="m="+n)

plt.title('No of deaths')    
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["infecteds"], label="m="+n)

plt.title('No of infected cells')    
plt.legend()
plt.show()
```
