```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

names = ["data", "q01", "q02", "q03", "q05"]
labels = ["none", "10%", "20%", "30%", "50%"]
source = []
for name in names:
    source.append(pd.read_csv("data/epidemic-" + name + ".csv"))
a = np.array(list(range(0,len(source[0]["infections"]))))
x = a.astype(float)

```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["infections"], label="q="+n)

plt.title('No of infections')     
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["deaths"], label="q="+n)

plt.title('No of deaths')    
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["infecteds"], label="q="+n)

plt.title('No of infected cells')    
plt.legend()
plt.show()
```
